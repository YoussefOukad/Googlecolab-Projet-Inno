```python
pseudo=[]
for x in caisse["Date"]:
  pseudo.append(x.date())
pseudo
```

```python
caisse1 = caisse.assign(dateonly=pseudo)
caisse1
```

```python
caissesorted=caisse1.sort_values('Date')
caissesorted
```

```python
A=[] #Liste de toutes les operation dans l'agence de zenith 
for x in caissesorted["Operation"]:
  if x not in A:
    A.append(x)
print(len(A), A)
```

```python
Dates=[]
for x in caissesorted['dateonly']:
  if x not in Dates:
    Dates.append(x)
Dates
```

```python
categoriesdates=[]
for x in Dates:
  s=0
  for y in caissesorted['dateonly']:
    if x==y:
      s+=1
  categoriesdates.append(s)
categoriesdates
```

```python
op=[]
for y in caissesorted["Operation"]:
  op.append(y)
op
```

```python
categories=[op[0:categoriesdates[0]]]
s=categoriesdates[0]
for i in range(len(categoriesdates)):
  if i!=len(categoriesdates)-1:
    t=s+categoriesdates[i+1]
    categories.append(op[s:t])
    s=t
categories
```

```python
opj=[]
for x in categories:
  f=[]
  for y in A: 
    s=0
    for z in x: 
      if y==z:
        s+=1
    f.append(s)
  opj.append(f)
opj 
#pour chaque jour on determine le nombre des operations enregistrees selon l'ordre donné par a liste A 
```

```python
A
```

```python
Dates
```

```python
ar = np.array(opj)
df = pd.DataFrame(ar, index = Dates, columns = A)
df
```

```python
pour=[]
for x in opj:
  e=[]
  for i in range(len(x)): 
    p=(x[i]/categoriesdates[i])*100     #c'est un pourcentage
    e.append(p)
  pour.append(e)
pour
```

```python
len(pour)
```

```python
pour[0][0]
```

```python
pourtot=[]
for i in range(len(A)):
  s=0
  for j in range(len(pour)):
    s+=pour[j][i]
  pourtot.append(s/len(pour))
pourtot
```

```python
A
```

```python
data1=[]
for i in range(len(pourtot)):
  for j in range(len(A)):
    if i==j :
      data1.append([A[i],pourtot[i]])
tabrecap=pd.DataFrame(data1, columns=['Operation', 'Pourcentage total'])
tabrecap
```

```python

tabrecap.plot(x="Operation", y="Pourcentage total", kind="bar")
```
