```python
Dates=[]
for x in pseudo:
  if x not in Dates:
    Dates.append(x)
Dates
```

```python
somme=[]
for x in Dates :
  s=0
  for y in pseudo :
    if x==y:
      s+=1
  somme.append(s)
somme
```

```python
categories=[a[0:somme[0]]]
s=somme[0]
for i in range(len(somme)):
  if i!=len(somme)-1:
    t=s+somme[i+1]
    categories.append(a[s:t])
    s=t
categories
```

```python
numpy=np
o=[]
for x in categories:
  L=[numpy.timedelta64(0,'ns')]
  for i in range(len(x)-1):
    e=x[i+1]-x[i] 
    if e < numpy.timedelta64(3600000000000,'ns'):
      L.append(e) 
  o.append(L)
o
```

```python
H=[]
for y in o:
  c=[]
  for x in y:
    c.append(x.astype('timedelta64[m]'))
  H.append(c)
H # liste de liste des durees entre deux enregistrements successives selon la journee
```

```python
moyjour=[]
for x in H:
  s=0
  for y in x:
    s+=y
  moyjour.append(s/len(x))
moyjour
```

```python
s=0
for x in moyjour:
  s+=x
moyenne=s/len(moyjour)
moyenne
```

```python
W=[versement, rem_check, autre_op_agence, remise_lcn, virement, retrait_bord, frais, retrait_cheque, retrait_carnet, non_categorise, mad, restitution, certification, opposition, blocage, recharge, dgi, achat_billet, vente_billet, tgr, retrait_mad]
MOY=[]
for T in W:
    a=T['Date'].values
    pseudo=[]
    for x in T["Date"]:
        pseudo.append(x.date())
    Dates=[]
    for x in pseudo:
        if x not in Dates:
            Dates.append(x)
    somme=[]
    for x in Dates :
        s=0
        for y in pseudo :
            if x==y:
                s+=1
        somme.append(s)
    categories=[a[0:somme[0]]]
    s=somme[0]
    for i in range(len(somme)):
        if i!=len(somme)-1:
            t=s+somme[i+1]
            categories.append(a[s:t])
            s=t
    numpy=np
    o=[]
    for x in categories:
        L=[numpy.timedelta64(0,'ns')]
        for i in range(len(x)-1):
          e=x[i+1]-x[i] 
          if e < numpy.timedelta64(3600000000000,'ns'):
            L.append(e) 
        o.append(L)
    H=[]
    for y in o:
        c=[]
        for x in y:
            c.append(x.astype('timedelta64[m]'))
        H.append(c)
    moyjour=[]
    for x in H:
        s=0
        for y in x:
            s+=y
        moyjour.append(s/len(x))
    s=0
    for x in moyjour:
        s+=x
    moyenne=s/len(moyjour)
    MOY.append(moyenne)
MOY
```

```python
Acommerce=[] #Liste de toutes les operations commerciales dans l'agence de zenith 
for x in commerce["Operation"]:
  if x not in Acommerce:
    Acommerce.append(x)
Acommerce
```

```python
Modification=(commerce.loc[(commerce["Operation"]=="Modification"), ["Date","Operation"]]).sort_values("Date")
Vente=(commerce.loc[(commerce["Operation"]=="Vente"), ["Date","Operation"]]).sort_values("Date")
Ouverture=(commerce.loc[(commerce["Operation"]=="Ouverture"), ["Date","Operation"]]).sort_values("Date")
```

```python
W=[Modification, Vente, Ouverture]
MOYcommerce=[]
for T in W:
    a=T['Date'].values
    pseudo=[]
    for x in T["Date"]:
        pseudo.append(x.date())
    Dates=[]
    for x in pseudo:
        if x not in Dates:
            Dates.append(x)
    somme=[]
    for x in Dates :
        s=0
        for y in pseudo :
            if x==y:
                s+=1
        somme.append(s)
    categories=[a[0:somme[0]]]
    s=somme[0]
    for i in range(len(somme)):
        if i!=len(somme)-1:
            t=s+somme[i+1]
            categories.append(a[s:t])
            s=t
    numpy=np
    o=[]
    for x in categories:
        L=[numpy.timedelta64(0,'ns')]
        for i in range(len(x)-1):
          e=x[i+1]-x[i] 
          if e < numpy.timedelta64(3600000000000,'ns'):
            L.append(e) 
        o.append(L)
    H=[]
    for y in o:
        c=[]
        for x in y:
            c.append(x.astype('timedelta64[m]'))
        H.append(c)
    moyjour=[]
    for x in H:
        s=0
        for y in x:
            s+=y
        moyjour.append(s/len(x))
    s=0
    for x in moyjour:
        s+=x
    moyenne=s/len(moyjour)
    MOYcommerce.append(moyenne)
MOYcommerce
```
