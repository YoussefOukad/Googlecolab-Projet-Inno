```python
W=[Modification, Vente, Ouverture]
MOYcommerce=[]
for T in W:
    a=T['Date'].values
    pseudo=[]
    for x in T["Date"]:
        pseudo.append(x.date())
    Dates=[]
    for x in pseudo:
        if x not in Dates:
            Dates.append(x)
    somme=[]
    for x in Dates :
        s=0
        for y in pseudo :
            if x==y:
                s+=1
        somme.append(s)
    categories=[a[0:somme[0]]]
    s=somme[0]
    for i in range(len(somme)):
        if i!=len(somme)-1:
            t=s+somme[i+1]
            categories.append(a[s:t])
            s=t
    numpy=np
    o=[]
    for x in categories:
        L=[numpy.timedelta64(0,'ns')]
        for i in range(len(x)-1):
          e=x[i+1]-x[i] 
          if e < numpy.timedelta64(3600000000000,'ns'):
            L.append(e) 
        o.append(L)
    H=[]
    for y in o:
        c=[]
        for x in y:
            c.append(x.astype('timedelta64[m]'))
        H.append(c)
    moyjour=[]
    for x in H:
        s=0
        for y in x:
            s+=y
        moyjour.append(s/len(x))
    s=0
    for x in moyjour:
        s+=x
    moyenne=s/len(moyjour)
    MOYcommerce.append(moyenne)
MOYcommerce
```
